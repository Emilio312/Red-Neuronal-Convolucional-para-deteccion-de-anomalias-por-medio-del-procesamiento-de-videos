```python
from keras.models import load_model
import time
import numpy as np
import cv2
#import nframe as nf
from PIL import Image
from numpy.matlib import repmat
import math
%matplotlib inline
```

```python
import math

def msf(numero):
    parte_decimal, parte_entera = math.modf(numero)
    decimal = 100*(round(parte_decimal,3))
    frame = (parte_entera*60+decimal)*30
    return frame

msf(2.43)
```

```python
import math 
#Funcion que retorna la cantidad de segundos en un intervalo
def segs (inicio_video, final_video):
    decimal_inicial, entero_inicio = math.modf(inicio_video) 
    decimal_final, entero_final = math.modf(final_video) 
    decimal_inicial =  100*(round(decimal_inicial,3))
    decimal_final =  100*(round(decimal_final,3))

    ef = entero_final
    ei = entero_inicio
    df = decimal_final
    di = decimal_inicial
    
    segs = 60*(ef-ei) + (df-di)
    return int(segs)

segs(3,4)
```

```python
#Cargar modelo
model = load_model('model_1C128N0.5DROPBATCH64, \n
20Ssteps35epochs_data2.2.1.h5')
```

```python
#Declaración de variables 
videopath = 'DataSernapesca//Con Bacterias//120124//13-03-2019//T1.avi'
cap = cv2.VideoCapture(videopath)
fps = cap.get(cv2.CAP_PROP_FPS)
cont = 1
cont1 = 0
inicio_video = 3.50
final_video = 9.45
seg = segs(inicio_video, final_video)
fpsextr = 3
a = np.zeros((seg*fpsextr, 224,224,3))
print(seg)
```

```python
tiempos = []
now = time.time()
while cap.isOpened():
    ret, frame = cap.read()
    
    if ret == False:
        break
        
    if(msf(inicio_video) <= cont <= msf(final_video)):
        if(cont%(30//fpsextr) == 0):
            image = np.array(frame)
            image.resize(1,224,224,3)
            
            if (cont1 < seg*fpsextr):
                a[cont1,:,:,:] = image
                cont1+=1
            
            elif (cont1 == seg*fpsextr):
                cont1 = 0
    
            duration = cont/fps
            minutes = int(duration/60)
            seconds = int(duration%60)
            tiempos.append(str(minutes) + ':' + str(seconds))
    
    cont = cont+1         
    
    if (cont == (msf(final_video)+1)):
        break
cap.release

print(f'Duración evaluación del video :\n
      {(time.time()-now)/60} Minutos Aprox.')
```

```python
now = time.time()
result = model.predict(a)
print(f'Duración predicción de las imágenes :\n
{(time.time()-now)/60} Minutos Aprox.')
```

```python
#Resultados bacterias sobre 0.5
resultados_b = (result>0.5)*1
```

```python
resultados_b[:10]
```

```python
frec_intervalo = 10
tomador_decision = 2

resultados_f = np.zeros(resultados_b.shape[0]-frec_intervalo)

for i in range (0, resultados_b.shape[0]-frec_intervalo, 1):
    temp = resultados_b[i:i+frec_intervalo]
    resultados_f[i] = temp.sum() > tomador_decision

print(f'Intervalos de tiempo con bacterias :\n
{((resultados_f == 1)*1).sum()}')

print(f'Intervalos de tiempo sin bacterias :\n
{((resultados_f == 0)*1).sum()}')
```

```python
resultados_f
```

```python
intervalos = []
i = 0 
while (i < resultados_f.shape[0]):
    if resultados_f[i] == 1:
        inic = i
        i+=1    
            
        while resultados_f[i] == 1:
                i+=1
                if i == resultados_f.shape[0]:
                    break 
        fin = i-1
        intervalos.append((inic,fin))
    
    i+=1
print(intervalos)
```

```python
for n in range (0, len(intervalos)):
    print(f'duración\n
:{tiempos[intervalos[n][0]]}-{tiempos[intervalos[n][1]]}')
```
